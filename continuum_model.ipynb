```python
import model_builder as mb
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Example: Define a custom kpath from Gamma to M to K to Gamma with default 20 points
def custom_kpath_generator(num_points=20):
    """
    Generate a custom fractional kpath from Gamma to M to K to Gamma.
    
    Parameters:
    num_points (int): Number of points in each segment of the kpath.

    Returns:
    kpath (ndarray): 2D array of fractional kpoints along the path.
    high_symmetry_kpoints (list of int): List of indices of high symmetry kpoints.
    high_symmetry_labels (list of str): List of labels for the high symmetry kpoints.
    """
    Gamma = np.array([0.0, 0.0])
    M = np.array([0.5, 0.0])
    K = np.array([1/3, 1/3])

    path1 = np.linspace(Gamma, M, num_points, endpoint=False)
    path2 = np.linspace(M, K, num_points, endpoint=False)
    path3 = np.linspace(K, Gamma, num_points+1, endpoint=True)
    
    kpath = np.vstack((path1, path2, path3))
    high_symmetry_kpoints = [0, num_points, 2*num_points, 3*num_points]
    high_symmetry_labels = [r'$\Gamma_M$', r'$M_M$', r'$K_M$', r'$\Gamma_M$']
    
    return kpath, high_symmetry_kpoints, high_symmetry_labels

kpath, high_symmetry_kpoints, high_symmetry_labels = custom_kpath_generator(20)
```

### Full Model for tMoTe2

```python
# Example 1: Full Model for tMoTe2

# ====================== Configuration Parameters ======================

material = "mote2_standard_basis"    # Name of the material (e.g., "mote2_standard_basis", "wse2_quick_basis","mote2_quick_basis")
angle = '2.13'                       # Angle parameter as a string (e.g., '2.13')
model_type = 'full'                  # Type of model (e.g., 'full', 'reduced')
n_jobs = -1                          # Number of parallel jobs (set to -1 to use all available cores)
path_model = "./"                    # Base path to the model data

# ====================== End of Configuration ======================


model = mb.ContinuousModel(
    material=material,
    angle=angle,
    model_type=model_type,
    path_model=path_model,
    kpath=kpath
)

# Build Hamiltonians for all k-points
model.build_hamiltonians(n_jobs=n_jobs, show_progress=False)  # show_progress=True to display a progress bar (not recommended for Jupyter, but works in Python scripts)

# Diagonalize all Hamiltonians to obtain eigenvalues and eigenvectors
model.diagonalize_hamiltonians(n_jobs=n_jobs, show_progress=False) # show_progress=True to display a progress bar (not recommended for Jupyter, but works in Python scripts)

# Plot the band structure
model.plot_band_structure(high_symmetry_kpoints, high_symmetry_labels)
```
